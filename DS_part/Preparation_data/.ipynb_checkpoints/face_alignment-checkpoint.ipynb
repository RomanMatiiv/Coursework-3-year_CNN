```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline

PATH_TO_CASCADE='/Users/roman/anaconda3/pkgs/libopencv-3.4.1-he076b03_1/share/OpenCV/haarcascades/haarcascade_eye.xml'
```

```python
img=cv.imread("/Users/roman/Desktop/3473_2.jpg")
img=cv.cvtColor(img, cv.COLOR_BGR2RGBA)
# img=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray");
```

### Найти глаза

```python
plt.imshow(img,cmap="gray");
```

```python
eye_cascade = cv.CascadeClassifier(PATH_TO_CASCADE)
eyes = eye_cascade.detectMultiScale(img)
```

```python
img_copy=img.copy()
for (ex,ey,ew,eh) in eyes:
        cv.rectangle(img_copy,(ex,ey),(ex+ew,ey+eh),(255,0,255),2)
```

```python
plt.imshow(img_copy);
```

### Понять на сколько нужно повернуть

```python
center=[]
for (ex,ey,ew,eh) in eyes:
    xc = int(np.mean([ex, ex+ew]))
    yc = int(np.mean([ey, ey+eh]))
    center.append([xc,yc])
    
center=np.array(center)

min_x=np.argmin(center,axis=0)
max_x=np.argmax(center,axis=0)

a=center[:,1].max()-center[:,1].min()
b=center[:,0].max()-center[:,0].min()
c=np.sqrt(a**2+b**2)

rad=np.arcsin(a/c)
corner=np.rad2deg(rad)


if center[min_x[0]][1]>=center[max_x[0]][1]:
    corner=-corner

rows,cols = img.shape[:2]

img_copy=img.copy()
M = cv.getRotationMatrix2D((cols/2,rows/2),corner,1)
img_copy = cv.warpAffine(img_copy,M,(cols,rows))
plt.imshow(img_copy,cmap="gray");
```
